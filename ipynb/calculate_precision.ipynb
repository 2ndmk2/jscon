```python
import glob
import numpy as np
import matplotlib.pyplot as plt
from jscon import astrometry
import importlib 
importlib.reload(astrometry)
from jscon import make_image as mk
from astropy.coordinates import SkyCoord  # High-level coordinates
import astropy.units as u
import os
from jscon import parameter_load
from jscon import analyze_result
importlib.reload(analyze_result)
import matplotlib as mpl
label_size = 14
mpl.rcParams['xtick.labelsize'] = label_size 
mpl.rcParams['ytick.labelsize'] = label_size 

def compute_precision(dx_dif):
    precision= np.percentile(dx_dif, [16, 50, 84])
    #print("%f, %f, %f" %( precision[2], precision[1],  precision[0]))
    return  precision
    
fig_out = "./fig_precision"
```

```python
pix_arcsec_arr = np.arange(0.3, 1.0, 0.1)
fig = plt.figure(figsize = (7, 5))
ax1 = fig.add_subplot(111)
n_star_arr = []
for (i, pix_scale) in enumerate(pix_arcsec_arr):
    out_folder = "../result/pixel_change/out%.1f" % pix_scale
    para_file = "/home/aizwms/jasmine/jasmin_git/jscon/params/pixel_change/para_%.1f.dat" % pix_scale
    dx_dif, dy_dif, dx_arr, dy_arr = analyze_result.analyze_result(out_folder, para_file, n_star = 100, plot = False)
    precision = compute_precision(dx_dif)
    plt.scatter(pix_scale, precision[1], color="k")
    print(pix_scale, precision)
    n_star_arr.append(len(dx_arr[0]))
#plt.yscale("log")
plt.xlabel("pix scale (arcsec)", fontsize = 18)
plt.ylabel("preicision (mas)", fontsize = 18)
plt.ylim(0, 18)
plt.axhline(y=4)# xmin=0, xmax=1, hold=None, **kwargs)
plt.savefig(os.path.join(fig_out , "precision_change_dpix.png"), bbox_inches='tight')
plt.title("Image size = $1800\\times1000$", fontsize = 15)
plt.show()
```

```python
fig = plt.figure(figsize = (7, 5))
plt.scatter(pix_arcsec_arr,n_star_arr)
plt.xlabel("pix scale (arcsec)", fontsize = 18)
plt.ylabel("N star", fontsize = 18)
plt.savefig(os.path.join(fig_out ,"./num_star_change_dpix.png"), bbox_inches='tight')
plt.show()
```

```python
pix_arcsec_arr = np.arange(0.9, 0.2, -0.1)
fig = plt.figure(figsize = (7, 5))
ax1 = fig.add_subplot(111)
n_star_arr = []
for (i, pix_scale) in enumerate(pix_arcsec_arr):
    file = "../result/pixel_change/out%.1f/input_astro_info.npz" % pix_scale
    astro_info = np.load(file)
    mask_targets = astro_info["mask_targets"]
    ra = astro_info["ra"][mask_targets] * 180/np.pi
    dec = astro_info["dec"][mask_targets]    * 180/np.pi
    c = SkyCoord(ra, dec, frame="icrs", unit="deg")
    gal_l = c.galactic.l.deg
    gal_b = c.galactic.b.deg
    plt.scatter(gal_l, gal_b, label ="dpix = %.1f asec" % pix_scale)

print(astro_info.files)
plt.legend()
plt.xlabel("gal l + 180 (deg)", fontsize = 18)
plt.ylabel("gal b", fontsize = 18)
plt.savefig(os.path.join(fig_out ,"./precision_field.png"), bbox_inches='tight')
plt.title("Image size = $1800\\times1000$", fontsize = 15)
plt.show()
```
